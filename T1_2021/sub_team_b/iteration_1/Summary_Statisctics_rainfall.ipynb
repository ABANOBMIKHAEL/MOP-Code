## Data visualization

```python
df.groupby(['day of week'])['Rainfall amount (millimetres)'].sum()
```

```python
#insepct col Rainfall amount group by day
values = df.groupby(['day of week'])['Rainfall amount (millimetres)'].sum().values
labels = df.groupby(['day of week'])['Rainfall amount (millimetres)'].sum().index


plt.figure(1, figsize=(6,6))
plt.style.use('seaborn')
explode = [0.0, 0, 0.1 , 0, 0, 0, 0.2]
plt.pie(
       values, 
       labels = labels,
       shadow = True,
       autopct= '%0.2f%%',
       explode = explode,
       textprops={'fontsize':12.5}
)

plt.title('The amount of rainfall group by day_week', fontsize =14)
plt.show()
```

```python
plt.figure(2)
sns.barplot(x = 'day of week', y='Rainfall amount (millimetres)',data =df,estimator= sum)

plt.title('which day of week has more rainfall')
plt.tight_layout()
plt.show()
```

```python
#insepct col Rainfall amount group by month

values_month = df.groupby(['Month'])['Rainfall amount (millimetres)'].sum().values
labels_month = df.groupby(['Month'])['Rainfall amount (millimetres)'].sum().index
explode_2 = [0, 0, 0, 0.2, 0.2, 0, 0, 0, 0, 0, 0, 0]
plt.figure(figsize= (12,8))
plt.pie(
       values_month, 
       labels = labels_month,
       shadow = True,
       autopct= '%0.2f%%',
       explode = explode_2 ,
       textprops={'fontsize':12})
#plt.legend(loc = 'upper right')
plt.show()
#ax1.legend(loc = 0)
```

```python
sns.barplot(x= 'Month', y= 'Rainfall amount (millimetres)', data = df, estimator=sum)
plt.title('Which mont has more rainfall')
plt.tight_layout()
plt.show()
```

```python
#insepct col Rainfall amount groub by year

values_year = df.groupby(['Year'])['Rainfall amount (millimetres)'].sum().values
labels_year = df.groupby(['Year'])['Rainfall amount (millimetres)'].sum().index


plt.figure(1)
explode_3 = [0, 0, 0, 0, 0, 0.2,0]
plt.pie(
       values_year, 
       labels = labels_year,
       shadow = True,
       autopct= '%0.2f%%',
       explode = explode_3,
       textprops= {'fontsize':12}
)
plt.title('pie chart group by year', fontsize =14)
plt.show()
#ax1.legend(loc = 1)
plt.clf()

plt.figure(2)
sns.barplot(x = labels_year, y=values_year)
plt.tight_layout()
plt.title('which year has more rainfall', fontsize = 15)
plt.show()

```

## Statistics for amount rainfall  

```python
df['Rainfall amount (millimetres)'].describe()
```

```python
rainfall_amount = df['Rainfall amount (millimetres)'].values
#hist
plt.figure(figsize=(13,6))
plt.style.use('seaborn')
plt.subplot(1,2,1)

plt.hist(rainfall_amount,
        range = (0,50),
        bins = 30,
        edgecolor = 'black')

plt.axvline(np.quantile(rainfall_amount, 0.5), c = 'black', label= 'median')
plt.axvline(np.mean(rainfall_amount), c = 'r', label ='mean')
plt.axvline(np.amax(rainfall_amount), c = 'green', label ='max')
#plt.axvline(np.amin(rainfall_amount), c = 'green', label ='min')

plt.title('Histogram of rainfall', fontsize= 14)
plt.legend( loc = 'upper right')

#log
plt.subplot(1,2,2)
log_rainfall_amout = np.log(rainfall_amount)
plt.hist(log_rainfall_amout,
        range = (0,50),
        bins = 40,
        edgecolor = 'black')
plt.title('(LOG)Histogram of rainfall', fontsize= 14)

plt.show()
```

```python
from scipy import stats
print('before log:', stats.describe(rainfall_amount))

print('\nAfter log:', stats.describe(log_rainfall_amout))
```

```python
# boxplot
plt.figure(2)
plt.boxplot(rainfall_amount)
plt.title('Box plot of rainfall', fontsize= 20)
plt.show()

print('median',np.median(rainfall_amount))
print('mean',np.average(rainfall_amount))
print('max',np.amax(rainfall_amount))
print('min',np.amin(rainfall_amount))
```
