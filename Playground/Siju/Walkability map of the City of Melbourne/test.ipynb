## new success 获取人口普查 gov.au 来源

```python
# 替换以下URL为您的Excel文件链接
excel_url = 'https://www.abs.gov.au/statistics/people/population/regional-population/2021-22/32180DS0001_2021-22r.xlsx'

# 指定需要读取的工作表名或索引
sheet_name = 'Table 2'

# 读取包含表头的数据行
header_df = pd.read_excel(excel_url, sheet_name=sheet_name, header=None, nrows=8)

# 提取并合并表头行
headers = header_df.iloc[5:8]
combined_header = headers.apply(lambda x: ' '.join(x.dropna().astype(str)), axis=0)

# 读取实际数据，并应用合并后的表头
table_2_df = pd.read_excel(excel_url, sheet_name=sheet_name, skiprows=8, header=None)
table_2_df.columns = combined_header
```

```python

# 显示数据
print("数据读取并手动合并表头成功。以下是前几行数据：")
display(table_2_df.head())
```

```python
# 过滤数据
st_code_col = 'S/T code'
filtered_df = table_2_df[table_2_df[st_code_col] == 2].dropna(subset=[st_code_col])
```

```python

# 显示过滤后的数据
print("过滤后的数据预览：")
display(filtered_df.head())
```

```python
# 假设 'SA3 code' 是列名中的一个，您需要根据实际情况调整
sa3_code_col = 'SA3 code'

# 过滤出所有 'SA3 code' 等于 20604 的行
filtered_sa3_df = table_2_df[table_2_df[sa3_code_col] == 20604]

# 显示过滤后的数据
print("SA3 code 为 20604 的数据预览：")
display(filtered_sa3_df.head())
```

```python
# 创建 DataFrame 的副本
filtered_sa3_df_copy = filtered_sa3_df.copy()
```

```python
# 计算每个区域的总人口
# 假设 'Population density 2022 persons/km2' 是人口密度列名
# 'Area km2' 是区域面积列名
filtered_sa3_df_copy.loc[:, 'Total Population'] = (
    filtered_sa3_df_copy['Population density 2022 persons/km2'] * filtered_sa3_df_copy['Area km2']
)

# 显示计算后的数据
print(filtered_sa3_df_copy[['SA2 name', 'Total Population']].head())
```

```python
filtered_sa3_df_copy
```

```python
# 列出要删除的列名
columns_to_drop = ['S/T code', 'S/T name', 'GCCSA code', 'GCCSA name', 'SA4 code', 'SA4 name', 'SA3 code', 'SA3 name']

# 删除这些列
filtered_sa3_df_copy.drop(columns=columns_to_drop, axis=1, inplace=True)

# 查看删除列后的DataFrame
filtered_sa3_df_copy.head()
```

## 获取LGA在线文件

```python
import requests
import geopandas as gpd
import tempfile
import os
from zipfile import ZipFile
from io import BytesIO

# 在线ZIP文件的URL
zip_url = 'https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files/SA2_2021_AUST_SHP_GDA2020.zip'

# 发送请求获取ZIP文件的内容
response = requests.get(zip_url)
zip_file_bytes = BytesIO(response.content)

# 创建一个临时目录来存储解压的文件
with tempfile.TemporaryDirectory() as temp_dir:
    # 解压ZIP文件
    with ZipFile(zip_file_bytes) as zip_file:
        zip_file.extractall(temp_dir)

    # 获取.shp文件的完整路径
    shp_file_name = [f for f in os.listdir(temp_dir) if f.endswith('.shp')][0]
    full_shp_path = os.path.join(temp_dir, shp_file_name)

    # 读取Shapefile到新的GeoDataFrame
    LGAgdf = gpd.read_file(full_shp_path)

    # 显示数据
    print(LGAgdf.head())
```

```python
LGAgdf
```

```python
# 提取 LGAgdf 中 'SA3_CODE21' 列为 20604 的行
filtered_LGAgdf = LGAgdf[LGAgdf['SA3_CODE21'] == 20604]

# 显示过滤后的数据
print(filtered_LGAgdf.head())
```

```python
# 检查 'SA3_CODE21' 列的数据类型
print(LGAgdf['SA3_CODE21'].dtype)

# 如果需要，将 'SA3_CODE21' 列转换为整数类型
LGAgdf['SA3_CODE21'] = pd.to_numeric(LGAgdf['SA3_CODE21'], errors='coerce')

# 再次尝试过滤
filtered_LGAgdf = LGAgdf[LGAgdf['SA3_CODE21'] == 20604]

# 显示过滤后的数据
print(filtered_LGAgdf.head())
```

## 获取人口和住在的地区列表和LAG 三者的区别

```python
# 人口
# 获取 'SA2 name' 列中所有不同的值
sa2_name_col = 'SA2 name' 
unique_sa2_names = filtered_sa3_df_copy[sa2_name_col].unique()

# 按照首字母排序
sorted_unique_sa2_names = sorted(unique_sa2_names)

# 计算不同值的数量
unique_sa2_names_count = filtered_sa3_df_copy[sa2_name_col].nunique()

print(f"'SA2 name' 列中的所有不同值：")
for name in sorted_unique_sa2_names:
    print(name)
print(f"'SA2 name' 列中不同值的数量为：{unique_sa2_names_count}")
```

```python
# 房屋
# 获取 'CLUE small area' 列中所有不同的值
clue_small_area_col = 'CLUE small area'  
unique_clue_small_areas = RESgdf[clue_small_area_col].unique()

# 按照首字母排序
sorted_unique_clue_small_areas = sorted(unique_clue_small_areas)

# 计算不同值的数量
unique_clue_small_areas_count = RESgdf[clue_small_area_col].nunique()

print("这些值包括：")
for area in sorted_unique_clue_small_areas:
    print(area)
print(f"'CLUE small area' 列中不同值的数量为：{unique_clue_small_areas_count}")
```

```python
# LGALGA
# 获取'SA2_NAME21' 列的所有不同值
LGAsa2_name_col = 'SA2_NAME21'  
unique_LGAsa2_names = filtered_LGAgdf[LGAsa2_name_col].unique()

# 按照首字母排序
sorted_unique_LGAsa2_names = sorted(unique_LGAsa2_names) 

# 计算不同值的数量
unique_LGAsa2_name_col = filtered_LGAgdf[LGAsa2_name_col].nunique()

print("这些值包括：")
for area in sorted_unique_LGAsa2_names:
    print(area)
print(f"'CLUE small area' 列中不同值的数量为：{unique_LGAsa2_name_col}")
```

```python
# 创建人口数据的SA2 name集合
sa2_names_population = set(sorted_unique_sa2_names)

# 创建LGA数据的SA2_NAME21集合
sa2_names_lga = set(sorted_unique_LGAsa2_names)

# 找出共有的值（交集）
common_values = sa2_names_population.intersection(sa2_names_lga)

# 找出人口数据独有的值
unique_in_population = sa2_names_population.difference(sa2_names_lga)

# 找出LGA数据独有的值
unique_in_lga = sa2_names_lga.difference(sa2_names_population)

# 打印结果
print("共有的值：", common_values)
print("人口数据中独有的值：", unique_in_population)
print("LGA数据中独有的值：", unique_in_lga)
```

```python
# 将两列的值转换为集合
sa2_name_set = set(sorted_unique_sa2_names)
clue_small_area_set = set(sorted_unique_clue_small_areas)

# 找出两列中共同的值（交集）
common_values = sa2_name_set.intersection(clue_small_area_set)

# 找出仅在 'SA2 name' 中存在的值
unique_in_sa2_name = sa2_name_set.difference(clue_small_area_set)

# 找出仅在 人口数据 中存在的值
unique_in_clue_small_area = clue_small_area_set.difference(sa2_name_set)

# 打印结果
print(f"共同的值：{common_values}")
print(f"住宅数据中独有的值：{unique_in_sa2_name}")
print(f"人口数据中独有的值：{unique_in_clue_small_area}")
```

## 画图

```python
min_population = merged_data['Total Population'].min()
max_population = merged_data['Total Population'].max()
```

```python
import branca.colormap as cm

colormap = cm.LinearColormap(
    colors=['#808080', 'red'],
    vmin=min_population,
    vmax=max_population
)

```

```python
# 定义鼠标悬停时的高亮样式
def highlight_function(feature):
    return {
        'fillColor': '#ffaf00',  # 高亮颜色
        'color': 'black',
        'weight': 3,
        'fillOpacity': 0.7
    }
```

```python
import folium

# 创建以墨尔本为中心的地图
pop_map = folium.Map(location=[-37.8155, 144.95], tiles='cartodbpositron', min_zoom=11, zoom_start=13)

for _, row in merged_data.iterrows():
    # 根据人口数计算颜色
    color = colormap(row['Total Population'])

    # 为每个区域添加GeoJSON图层
    folium.GeoJson(
        row['geometry'],
        style_function=lambda feature, color=color: {
            'fillColor': color,
            'color': 'black',
            'weight': 0.5,
            'fillOpacity': 0.5  # 调整透明度
        },
        highlight_function=highlight_function,
        tooltip=f'SA2: {row["SA2 name"]}<br>Population: {row["Total Population"]}'
    ).add_to(pop_map)


```
