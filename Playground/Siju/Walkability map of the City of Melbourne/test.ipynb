```python
# 替换以下URL为您的Excel文件链接
excel_url = 'https://www.abs.gov.au/statistics/people/population/regional-population/2021-22/32180DS0001_2021-22r.xlsx'

# 指定需要读取的工作表名或索引
sheet_name = 'Table 2'

# 读取Excel文件，跳过前五行，将第六行到第八行作为多级表头
population_df = pd.read_excel(excel_url, sheet_name=sheet_name, header=None, skiprows=9)

# 查看数据以确定是否正确读取
print(population_df.head())
```

```python
try:
    # 读取包含表头的数据行
    header_df = pd.read_excel(excel_url, sheet_name=sheet_name, header=None, nrows=8)

    # 提取并合并表头行
    headers = header_df.iloc[5:8]
    combined_header = headers.apply(lambda x: ' '.join(x.dropna().astype(str)), axis=0)

    # 读取实际数据，并应用合并后的表头
    table_2_df = pd.read_excel(excel_url, sheet_name=sheet_name, skiprows=8, header=None)
    table_2_df.columns = combined_header
    
    print("数据读取并手动合并表头成功。以下是前几行数据：")
    display(table_2_df.head())
except Exception as e:
    print(f"手动合并表头时出错：{e}")
```

```python
# 'S/T code'是实际的列名
st_code_col = 'S/T code'

# 过滤掉'S/T code'列中不等于2的数据，包括空值
filtered_df = table_2_df[table_2_df[st_code_col] == 2].dropna(subset=[st_code_col])

# 显示过滤后的数据
print("过滤后的数据预览：")
print(filtered_df.head())
```
