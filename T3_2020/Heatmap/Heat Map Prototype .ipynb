```python
#pip install folium
import pandas as pd
import numpy as np
import folium
from folium import plugins

from sodapy import Socrata
```

```python
#folium.Map()
melb_city = folium.Map(location=[-37.80841814,144.95906317], zoom_start=14, width=1000, height=600, control_scale=True)
melb_city
```

```python
client = Socrata("data.melbourne.vic.gov.au", None)
sensor_locations = client.get("h57g-5234")
sensor_locations_df = pd.DataFrame.from_records(sensor_locations)
```

```python
sensor_locations_df
```

```python
# plot locations
for index, row in sensor_locations_df.iterrows():
    folium.Marker(location=[row['latitude'], row['longitude']], popup=row['sensor_name']).add_to(melb_city)
    #print([row['sensor_name'], row['sensor_name']])
```

```python
melb_city
```

```python
data = []
for index, row in sensor_locations_df.iterrows():
    #folium.Marker(location=[row['latitude'], row['longitude']], popup=row['sensor_name']).add_to(melb_city)
    data.append([row['latitude'], row['longitude'], np.random.random()])
```

```python
plugins.HeatMap(data).add_to(melb_city)
melb_city
```
