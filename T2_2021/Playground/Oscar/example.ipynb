# Constructing wordcloud

```python
import matplotlib.pyplot as plt #数据可视化
import jieba #词语切割
import wordcloud #分词
from wordcloud import WordCloud,ImageColorGenerator,STOPWORDS #词云，颜色生成器，停止词
import numpy as np #科学计算
from PIL import Image #处理图片
```

```python
import seaborn as sns
def ciyun():

    #with open('少年中国说.txt','r',encoding='gbk') as f:  #打开新的文本转码为gbk
    #    textfile= f.read()  #读取文本内容
    #wordlist = jieba.lcut(textfile)#切割词语
    #space_list = ' '.join(wordlist) #空格链接词语
    ##print(space_list)
    #
    table, client = deakin.top_table(top=20)



    str_data = (',').join(table.Name.tolist())
    wordlist = jieba.lcut(str_data)#切割词语
    space_list = ' '.join(wordlist)
    
#    space_list = "Oscar flower, Mirriam Resturant, albert gym, josh shop, jason's xxx, net's oooo, deep learning, machine learning, assignment, lockdown"*6
#    wordlist = jieba.lcut(space_list)#切割词语
#    space_list = ' '.join(wordlist)
    
    #backgroud = np.array(Image.open('map_1.png')) 
    backgroud = np.array(Image.open('/Users/wuyoscar/Desktop/CoM-Secondary_A_Black.jpg')) 

	
    wc = WordCloud(width=1200, height=2200,
			background_color='white',
	        mode='RGB', 
			mask=backgroud, #添加蒙版，生成指定形状的词云，并且词云图的颜色可从蒙版里提取
			max_words=500,
			stopwords=STOPWORDS.add('老年人'),#内置的屏蔽词,并添加自己设置的词语
			#font_path='C:\Windows\Fonts\STZHONGS.ttf',
			max_font_size=150,
			relative_scaling=0.6, #设置字体大小与词频的关联程度为0.4
			random_state=50, 
			scale=2 
			).generate(space_list) 
			
   # image_color = ImageColorGenerator(backgroud)#设置生成词云的颜色，如去掉这两行则字体为默认颜色
   # wc.recolor(color_func=image_color)
    fig,ax = plt.subplots(2,1,dpi=120,figsize=(10, 8))
    ax[0].imshow(wc) #显示词云
    ax[0].axis('off') #关闭x,y轴

    sns.barplot(ax=ax[1],y="Name",x ='downloadCount', data=table,palette="Blues_d")
    ax[1].set_title('Top %s datasets based on downloads.'%(20))
    plt.tight_layout()
    plt.show()
    plt.show()#显示
    #wc.to_file('map_1.jpg') #保存词云图

def main():
    ciyun()
 
if __name__ == '__main__':
    main()
table
```
