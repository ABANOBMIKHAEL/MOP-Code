```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import dataset from data.csv file
df = pd.read_csv('Dummy_Data.csv')
df.head()
```

```python
df.groupby(['DayOfWeek','Status']).size()
```

```python
df.groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(10, 6), rot=0)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Count of Unoccupied and Present Parkings", y=1.02)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
